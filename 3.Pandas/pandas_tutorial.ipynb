```python
import pandas as pd
data_sample = pd.read_excel("./datasets/data_simple.xlsx")
data_sample

# กรอง คนที่มีอายุมากกว่า10
data_year_more_than_10 = data_sample['age']>10
data_year_more_than_10
# อยากได้เป็น dataframe dataFrame[]
data_year_more_than_10 = data_sample[data_sample["age"]>10 ]
data_year_more_than_10

# เพิ่ม Column ชื่อว่า Sex
data_sample["sex"]=['Female', 'male', 'male', 'male', 'male', 'male']
data_sample

# เพิ่ม Column ชื่อว่า Weight
data_sample["Weight"]=[30,40,50,50,50,50]
data_sample

# กรองคนที่มีน้ำหนักมากกว่า 40
data_sample_weight_more_than_40 = data_sample["Weight"] > 40
data_sample_weight_more_than_40

# สร้าง dataFram ชื่อว่า data_sample_wieght_more_than_40
data_sample_weight_more_than_40 = data_sample[data_sample["Weight"] > 40]
data_sample_weight_more_than_40

# กรองคนที่มีน้ำหนักมากกว่า 40 และขอคนที่อายุมากกว่า 10 () & ()
(data_sample['Weight']>40) & (data_sample['age'] > 10)

# dataFrame คนที่มี น้ำหนักมากกว่า 40 และ อายุมากกว่า 10 () ฿ ()
data_sample_more_weigh_40_and_age_more_10 = data_sample[(data_sample['Weight']>40) & (data_sample['age'] > 10)]
data_sample_more_weigh_40_and_age_more_10

# กรองเพศชายอย่างเดียว
data_sample["sex"] == 'male'
data_sample["sex"] != 'male'

# สร้าง DataFrame ชื่อว่า data_sample_male
data_sample_female = data_sample[data_sample["sex"] != 'male']
data_sample_female

# DataFrame income น้อยกว่าหรีอเท่ากับ 30000 และ (&) น้ำหนักมากกว่าหรือเท่ากับ 30 หรือ(|) เพศชาย
data_sample_income = data_sample[(data_sample["income"]<=30000)& (data_sample["Weight"]>30) & (data_sample['sex'])]
data_sample_income



```
